```python
#libraries

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/online_retail_II.csv')
df.head()
```

```python
#Inspection

df.info()
df.describe()
df.shape
```

```python
#Rwemoving missing customerId

df = df.dropna(subset=['Customer ID'])
```

```python
# Removing negative Quantity

df = df[df['Quantity'] > 0]
```

```python
#Removing invalid prices

df = df[df['Price'] > 0]
```

```python
#Converting InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
#Reset index

df = df.reset_index(drop=True)
```

```python
#Total Amount

df['TotalAmount'] = df['Quantity'] * df['Price']
```

```python
#Year

df['Year'] = df['InvoiceDate'].dt.year
```

```python
#Month

df['Month'] = df['InvoiceDate'].dt.month
```

```python
#Year - Month

df['YearMonth'] = df['InvoiceDate'].dt.to_period('M').astype(str)
```

```python
#day of week

df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()
```

```python
#Validation

df.isnull().sum()
df.describe()
```

```python
#Removing Duplicates

df = df.drop_duplicates()
```

```python
#Top countries

df['Country'].value_counts().head()
```

```python
#Saving the clean data

df.to_csv("cleaned_retail.csv", index=False)
```
